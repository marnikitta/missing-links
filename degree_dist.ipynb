# Adamic/Adar metric distribution

```python
result = []
for dirName, subdirList, fileList in os.walk('adamic'):
    for fname in fileList:
        if (fname == 'part-00000'):
            with open(dirName + '/' + fname) as f:
                data = f.readlines()
                sample = np.random.choice(data, len(data) // 1000)
                for x in sample:
                    result.append(np.double(x.strip()))
result = np.array(result)
```

```python
len(result)
```

```python
plt.hist(np.log(result), log=True, bins = 100)
plt.show()
```

```python
plt.hist(result)
plt.show()
```
